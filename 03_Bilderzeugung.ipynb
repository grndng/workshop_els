# Digitale Bilder: Grundlagen

Dieses Notebook zeigt, dass ein Computer Bilder als **Pixel** speichert und wie wir eigene Bilder erzeugen k√∂nnen.

## 1. Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

## 2. Beispielbild laden (aus skimage, z.B. eine Astronautin)

```python
image = data.astronaut()        # RGB-Bild (H√∂he, Breite, 3)
print('Bildform:', image.shape)

plt.figure(figsize=(5, 5))
plt.imshow(image)
plt.title('Originalbild')
plt.axis('off')
plt.show()
```

## 3. Pixel untersuchen

Jedes Bild besteht aus Pixeln. Wir k√∂nnen uns einzelne Pixel und ihre Farbwerte anschauen.

```python
h, w, c = image.shape
print('H√∂he:', h, 'Breite:', w, 'Kan√§le:', c)

# Pixel in der Mitte ausgeben
center_pixel = image[h//2, w//2, :]
print('Pixel in der Mitte (RGB):', center_pixel)
```

## 4. Eigenes kleines Bild erzeugen

Hier erzeugen wir ein ganz kleines Bild (10√ó10 Pixel) mit einem roten Quadrat in der Mitte.

```python
# Bildgr√∂√üe
image_size = (10, 10, 3) # 10 x 10 Pixel gro√ües Bild, 3 B√§nder (Rot, Gr√ºn, Blau)
small_image = np.zeros(image_size, dtype=np.uint8)
small_image[2:8, 2:8, 0] = 255  # rotes Quadrat in der Mitte

plt.figure(figsize=(3, 3))
plt.imshow(small_image)
plt.title('Selbst erzeugtes Bild')
plt.show()
```
